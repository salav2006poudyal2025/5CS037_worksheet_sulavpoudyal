```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Dataset/medical_students_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
numeric_cols = ['Age','Height','Weight','BMI',
                'Temperature','Heart Rate','Blood Pressure','Cholesterol']

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())   # median → reduces effect of outliers

df['Student ID'] = df['Student ID'].fillna(-1)  # placeholder for unknown IDs
```

```python
cat_cols = ['Gender', 'Blood Type', 'Diabetes', 'Smoking']

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])   # mode → most common category
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
#  3.2 Exercises- Data Cleaning and Transformations with ”Titanic Dataset”:
#  Dataset Used: "titanic.csv"
#  Problem- 1:
#  Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.
#  Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,
#  median, maximum value, and minimum value of the ’Fare’ column
```

```python

import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Dataset/Titanic-Dataset.csv')

# 2) Subset the required columns
cols = ['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']
missing = [c for c in cols if c not in df.columns]
if missing:
    raise ValueError(f"Missing expected columns: {missing}")

subset = df[cols]

# 3) Filter to first-class passengers (Pclass == 1)
first_class = subset[subset['Pclass'] == 1].copy()

# Optional: handle non-numeric or missing 'Fare' values safely
first_class['Fare'] = pd.to_numeric(first_class['Fare'], errors='coerce')

#  4) Compute Fare statistics
fare_mean = first_class['Fare'].mean(skipna=True)
fare_median = first_class['Fare'].median(skipna=True)
fare_max = first_class['Fare'].max(skipna=True)
fare_min = first_class['Fare'].min(skipna=True)

#  5) Display results
print("First-class Fare statistics:")
print(f"Mean   : {fare_mean:.4f}")
print(f"Median : {fare_median:.4f}")
print(f"Max    : {fare_max:.4f}")
print(f"Min    : {fare_min:.4f}")

# If you also want to see a preview of the filtered DataFrame:
print("\nPreview of first-class subset:")
print(first_class.head(10))
```

```python
# Problem- 2:
#  How many null values are contained in the ’Age’ column in your subsetted DataFrame? Once you’ve found this out, drop them from your DataFrame.
```

```python

# Ensure Age is numeric to detect proper NaNs (handles strings like '' etc.)
first_class['Age'] = pd.to_numeric(first_class['Age'], errors='coerce')

#  1) Count nulls in Age
age_null_count = first_class['Age'].isna().sum()
total_rows = len(first_class)

print(f"Total first-class rows: {total_rows}")
print(f"Null values in 'Age': {age_null_count}")


#  2) Drop rows where Age is null
first_class_no_na_age = first_class.dropna(subset=['Age']).copy()

print(f"Rows after dropping missing 'Age': {len(first_class_no_na_age)}")

#quick check: confirm no null Age
assert first_class_no_na_age['Age'].isna().sum() == 0
```

```python
# Problem- 3:
#  The ’Embarked’ column in the Titanic dataset contains categorical data representing the ports of embarka
# tion:
#  • ’C’ for Cherbourg
#  • ’Q’ for Queenstown
#  • ’S’ for Southampton
#  Task:
#  1. Use one-hot encoding to convert the ’Embarked’ column into separate binary columns (’Embarked C’,
#  ’Embarked Q’, ’Embarked S’).
#  2. Add these new columns to the original DataFrame.
#  3. Drop the original ’Embarked’ column.
#  4. Print the first few rows of the modified DataFrame to verify the changes
```

```python
#  ensure 'Embarked' is treated as categorical/string
df['Embarked'] = df['Embarked'].astype(str).str.upper()

# 2) One-hot encode 'Embarked' into 'Embarked C', 'Embarked Q', 'Embarked S'
#  We use get_dummies with prefix='Embarked' and prefix_sep=' ' to match the exact column names requested.
embarked_dummies = pd.get_dummies(
    df['Embarked'],
    prefix='Embarked',
    prefix_sep=' ',
    dtype=int
)

# restrict to only C, Q, S (in case other values exist)
expected_cols = ['Embarked C', 'Embarked Q', 'Embarked S']
for col in expected_cols:
    if col not in embarked_dummies.columns:
        embarked_dummies[col] = 0  # add missing dummy columns if a category doesn't appear

# Reorder to C, Q, S
embarked_dummies = embarked_dummies[expected_cols]

# 3) Add new dummy columns to the original DataFrame
df = pd.concat([df, embarked_dummies], axis=1)

# 4) Drop the original 'Embarked' column
df = df.drop(columns=['Embarked'])

# 5) Print the first few rows to verify
print(df.head())
```

```python
#  Problem- 4:
#  Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column.
# Draw a visualization to show how the survival distributions vary by gender.
```

```python
import matplotlib.pyplot as plt
survival_rates = df.groupby("Sex")["Survived"].mean()
print("Mean survival rates by gender:")
print(survival_rates)

# Plot the survival rates
plt.figure(figsize=(6,4))
plt.bar(survival_rates.index, survival_rates.values)
plt.xlabel("Gender")
plt.ylabel("Mean Survival Rate")
plt.title("Survival Rate by Gender")
plt.tight_layout()
plt.show()
```
