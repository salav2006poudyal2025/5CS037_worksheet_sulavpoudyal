```python
import numpy as np

# 1. Empty array 2x2
arr_empty = np.empty((2,2))
print("Empty 2x2 array:\n", arr_empty)

# 2. All-one array 4x2
arr_ones = np.ones((4,2))
print("\nAll-one 4x2 array:\n", arr_ones)

# 3. Array filled with a given value
arr_full = np.full((3,3), 7)  # example fill value = 7
print("\nFull array 3x3 with value 7:\n", arr_full)

# 4. Zeros like another array
given = np.array([[1,2,3],[4,5,6]])
arr_zeros_like = np.zeros_like(given)
print("\nZeros like given array:\n", arr_zeros_like)

# 5. Ones like another array
arr_ones_like = np.ones_like(given)
print("\nOnes like given array:\n", arr_ones_like)

# 6. Convert list to numpy array
new_list = [1,2,3,4]
np_array = np.array(new_list)
print("\nConverted list to array:", np_array)
```

```python
# 1. Array 10 to 49
arr_range = np.arange(10, 50)
print("Array 10→49:\n", arr_range)

# 2. 3×3 matrix 0→8
matrix_3x3 = np.arange(9).reshape(3,3)
print("\n3x3 matrix 0→8:\n", matrix_3x3)

# 3. Identity matrix 3x3
identity = np.eye(3)
print("\nIdentity 3x3:\n", identity)

# 4. Random array 30 and mean
rand30 = np.random.random(30)
print("\nMean of random size-30 array:", rand30.mean())

# 5. Random 10×10 and min/max
rand10 = np.random.random((10,10))
print("\nMin:", rand10.min(), " Max:", rand10.max())

# 6. Zero array size 10, replace 5th element
z = np.zeros(10)
z[4] = 1   # 5th element index=4
print("\nZero array with 5th element = 1:\n", z)

# 7. Reverse array
arr = np.array([1,2,0,0,4,0])
rev = arr[::-1]
print("\nReversed array:", rev)

# 8. 2D array with 1 border and 0 inside
border = np.ones((5,5))
border[1:-1, 1:-1] = 0
print("\nBorder 1, inside 0:\n", border)

# 9. Checkerboard 8×8
checker = np.zeros((8,8), dtype=int)
checker[1::2, ::2] = 1
checker[::2, 1::2] = 1
print("\nCheckerboard 8x8:\n", checker)
```

```python
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

# 1. Addition
print("x + y:\n", x + y)

# 2. Subtraction
print("\nx - y:\n", x - y)

# 3. Multiply by integer (example: 3)
print("\n3 * x:\n", 3 * x)

# 4. Square elements
print("\nx squared:\n", x**2)

# 5. Dot products
print("\nv·w =", np.dot(v,w))
print("x·v =", np.dot(x, v))
print("x·y =\n", np.dot(x, y))

# 6. Concatenate along rows and columns
row_concat = np.vstack((x, y))
col_concat = np.column_stack((v, w))
print("\nRow concat x & y:\n", row_concat)
print("\nColumn concat v & w:\n", col_concat)

# 7. Concatenate x and v → why error?
try:
    np.concatenate((x, v))
except Exception as e:
    print("\nError when concatenating x and v:")
    print(e)
    print("\nExplanation: x is 2D (2×2) but v is 1D (size 2), so shapes do not match.")
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1. A * A^-1 = I
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("A * A^-1 =\n", I)

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB:\n", AB)
print("\nBA:\n", BA)
print("\nAB equals BA? →", np.allclose(AB, BA))

# 3. (AB)^T = B^T A^T
lhs = (AB).T
rhs = np.dot(B.T, A.T)
print("\n(AB)^T =\n", lhs)
print("B^T A^T =\n", rhs)
print("Equal? →", np.allclose(lhs, rhs))
```
